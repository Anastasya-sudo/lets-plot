```python
import pandas as pd
from lets_plot import *

LetsPlot.setup_html()
```

## Basic geom_ngon

```python
data = pd.DataFrame({
    'x': [0],
    'y': [0]
})

ggplot(data, aes('x', 'y')) + \
    geom_ngon(
        sidecount=5,
        size=70,
        fill='#f48067',
        color='black',
        alpha=0.8
    ) + \
    ggtitle('Basic geom_ngon')
```

## Sidecount parameter

```python
plots = []

for sides, x in [(3, 0), (4, 1.5), (5, 3), (6, 4.5)]:
    data = pd.DataFrame({
        'x': [x],
        'y': [0],
        'sides': [sides]
    })

    p = ggplot(data, aes('x', 'y')) + \
        geom_ngon(
            sidecount=sides,
            size=70,
            fill='#f48067',
            color='black',
            alpha=0.8,
            tooltips=layer_tooltips().line('sides|@sides')
        ) + \
        ggtitle(f'sides = {sides}')

    plots.append(p)

gggrid(plots, ncol=2)
```

## Fill aesthetic

```python
df_style = pd.DataFrame({
    'x': [1, 1.2, 1.4, 1.6],
    'y': [1, 1, 1, 1],
    'fill_color': ['#6E5160', '#8C4743', '#412227', '#CB4154']
})

ggplot(df_style, aes('x', 'y')) + \
    geom_ngon(
        aes(fill='fill_color'),
        size=70,
        sidecount=sides,
        color='black',
        stroke=1
    ) + \
    scale_fill_identity() +\
    ggtitle('fill aesthetic') +\
    ggsize(1800, 500) 
```

## Alpha aesthetic

```python
data = pd.DataFrame({
    'x': [1, 1.2, 1.4, 1.6],
    'y': [1, 1, 1, 1],
    'alpha_value': [0.25, 0.55, 0.75, 1.0]
})

ggplot(data, aes('x', 'y')) + \
    geom_ngon(
        aes(alpha='alpha_value'),
        sidecount=5,
        size=70,
        fill='#f48067',
        color='black'
    ) + \
    scale_alpha_identity() +\
    ggtitle('alpha aesthetic') +\
    ggsize(1800, 500)
```

## Size aesthetic

```python
data = pd.DataFrame({
    'x': [1, 1.3, 1.6, 1.9],
    'y': [1, 1, 1, 1],
    'size': [10, 30, 50, 70]
})

ggplot(data, aes('x', 'y')) + \
    geom_ngon(
        aes(size='size'),
        sidecount=5,
        fill='#5E2129',
        color='black',
        alpha = 0.8,
    ) + \
    ggtitle('size aesthetic') +\
    scale_size_identity() +\
    ggsize(1800, 500)
```

## Color aesthetic

```python
data = pd.DataFrame({
    'x': [1, 1.3, 1.6, 1.9],
    'y': [1, 1, 1, 1],
    'color': ['#C51D34', '#FF496C', '#8C4743', '#412227']
})

ggplot(data, aes('x', 'y')) + \
    geom_ngon(
        aes(color='color'),
        sidecount=5,
        fill='#5E2129',
        size=70,
        alpha = 0.3,
    ) + \
    ggtitle('color aesthetic') +\
    scale_color_identity() +\
    ggsize(1800, 500)
```

## Layer composition

```python
data = pd.DataFrame({
    'x': [0],
    'y': [0],
    'fill': ['#4F0014'],
    'color': ['#2f2f2f']
})

(
    ggplot(
        data,
        aes('x', 'y', fill='fill', color='color') 
    ) +
    geom_ngon(
        sidecount=4,
        size=80,
        alpha=0.3
    ) +
    geom_ngon(
        sidecount=9,
        size=50,
        alpha=0.8
    ) +
    scale_fill_identity() +
    scale_color_identity() +
    ggtitle('layer composition')
)
```
